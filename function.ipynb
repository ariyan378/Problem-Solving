```python
def is_even(num):#num = Parameter
    '''
    Docstring for is_even
    input : integer
    output : odd/even
    :param num: Description
    '''
    if type(num)==int:
        if num%2==0:
             return 'Even'
        else:
            return 'odd'
    else:
        return "Only integer Number is allowed" 

for i in range(1,6):
    x = is_even(i)

    print(x)#x = argument
```

```python
def multiply(*args):
    micro = 80

    for i in args:
        micro -= i
    
    print(args)
    return micro    
```

```python
multiply(2,2,3,4,6)#user can give any amount of value here without any tension 
```
